```python
import requests
import cv2
import numpy as np
import matplotlib.pyplot as plt

response = requests.get('https://i.imgur.com/c33hvpo.jpeg', headers = {'user-agent': 'utd-cs'})
img_array = np.asarray(bytearray(response.content))
img = cv2.cvtColor(cv2.imdecode(img_array, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
plt.imshow(img[:, :, 0], cmap='gray')
```

```python
plt.imshow(img[:, :, 2], cmap='gray')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img, cmap='gray')
```

```python
kernel = np.array([[-1,  0,  1],
                   [-2,  0,  2],
                   [-1,  0,  1]])

convolved_img = cv2.filter2D(img, -1, kernel)
plt.imshow(convolved_img, cmap='gray')
```
